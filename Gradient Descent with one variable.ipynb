```python
#Suppose we apply gradient descent to this function, starting with x = -3. To do this
#we will need to update x using the update rule:
#x = x ‚Äì alpha * derivative(x)
#where derivative is the derivative for our function f(x) above, and alpha is the
#‚Äústep size‚Äù. Setting x = -3 and alpha = 0.01, run gradient descent for 5 iterations
#(that is do the update 5 times). Give the values of x and f(x) a the start and after
#each of the first 5 iterations. If you have implement this correctly, the value of x
#after 1 iteration should be -2.01 and during the 5 iterations, the value of f(x) should
#be decreasing steadily.
```

```python
import pylab
import numpy
import math 


def f(x):
    return ((x**4)-(x**3)-6*(x**2)+2)

def derivative(x):
    h = 1./1000.
    rise = f(x + h) - f(x)
    run = h
    slope = rise / run
    return slope
```

```python
x = -3
for i in range (0,5):
    x = x - 0.01*derivative(x)
    print(x)
```

```python
#Run the gradient descent again, starting with x = -3 for 1000 iterations. Report the
#last 5 values of x and f(x). Has the value of x converged? Has the gradient descent
#found a minimum? Has it found the global minimum? 
```

```python
x = -3
for i in range (0,1000):
    x = x - 0.01*derivative(x)
    print(x)
```
